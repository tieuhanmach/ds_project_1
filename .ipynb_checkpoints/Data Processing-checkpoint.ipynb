Firstly I would like to look at verification method that hosts used

```python
# Get unique value of verification method
possible_vals=[]
ver=host['host_verifications']
for i in range(ver.shape[0]):
    for j in ver[i].split("'")[1::2]:
        if j not in possible_vals:
            possible_vals.append(j)  
```

```python
def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]

    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up
    '''
    new_df = {}
    #loop through list of ed types
    for val in look_for:
        new_df[val]=0
        #loop through rows
        for idx in range(df.shape[0]):
            #if the ed type is in the row add 1
            if val in df[col1][idx].split("'")[1::2] or val in df[col1][idx].replace('"','').replace('{','').replace('}','').split(","):
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values(col2, ascending=False, inplace=True)
    return new_df
```

```python
def clean_df(df, col1, possible_vals):
    count = df[col1].value_counts().reset_index()
    count.rename(columns={'index': col1, col1: 'count'}, inplace=True)
    count_df = total_count(count, col1, 'count', possible_vals)

    #Sum of price
    price = df.groupby([col1])['price'].sum().reset_index()
    price_df = total_count(price,  col1, 'price', possible_vals)

    props_df=pd.concat([count_df,price_df[['price']]],axis=1)
    props_df['props'] = props_df['count']/count_df['count'].sum()
    props_df['price_avg']=props_df['price']/props_df['count']
        
    return props_df
```

```python
props_method_df=clean_df(host, 'host_verifications', possible_vals)
props_method_df
```

```python
def plot(df,col1, col2, title):
    df_plot=df[[col1, col2]].set_index(col1).sort_values(by=col2)
    df_plot.plot(kind='bar', legend=None);
    plt.title(title);
    plt.show()
```

```python
# Plot the propotions of verification method
plot(props_method_df,'host_verifications','props', 'Verification method distribution')
```

> Phone, email, review are the most frequent verification method

```python
# Plot the average price of each method
plot(props_method_df,'host_verifications','price_avg', 'Price per method')
```

> In contrast, manual online, sent_id and linkedint are the methods that have highest price

```python
# Build model to predict price base on property numeric and cateorical variables:
# Define X,y
y=prop['price']
# Drop unnecessary variables:
df=prop.drop(['has_availability','calendar_updated_day','price','amenities'],axis=1)
# Dummy the categorical variables
cat_vars = df.select_dtypes(include=['object']).copy().columns
for var in cat_vars:
# for each cat add dummy var, drop original column
    df = pd.concat([df.drop(var, axis=1), pd.get_dummies(df[var], prefix=var, prefix_sep='_', drop_first=True)], axis=1)
X = df

```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=.3,random_state=42)
# Initiate
lm_model=LinearRegression(normalize=True)
#Fit
lm_model.fit(X_train,y_train)
y_pred=lm_model.predict(X_test)
r2_test=r2_score(y_test, y_pred)
r2_test
```
